```python
result1
```

```python
ebola_melt
```

```python
ebola_melt.drop(columns='variable',inplace=True)
```

```python
ebola_melt
```

```python
weather = pd.read_csv('data/weather.csv')
```

```python
weather_melt.head(4)
```

```python
weather_tidy = weather_melt.pivot_table(index=['id', 'year', 'month','day'],
                            columns='element',
                            values='temp')
```

```python
weather_tidy.reset_index(inplace=True)
```
